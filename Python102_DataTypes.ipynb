# Data Types

The basic data types - int, float, None, bool. They are also known as Scalar types (single value). We'll learn about Collection types later in Data Structures.

**Pro Tip**: Use type() to find out the data type. 

```python
print('Hello')

print(type(None))
print(type(17))
print(type(1.23))
print(type(True))
```
